# Test incident reports
## TIR1
* Problem : Device doesn't turn off when GPIO8 goes low.  The current circuitry doesn't turn off the power because the GPIO that controls the bottom anode of D1 is pulled  up.  The default state of this pin must be low (by using a pull-down resistor).  Currently GPIO8 is used, which must be pulled up at boot.
* Proposed fix : So let's swap HY2.3 (IO4) and HY2.7 (IO8).
   * IO8 becomes MISO (pulled up)
   * IO4 becomes PWR_EN
   * Add 10K pull down on IO4.
## TIR2
* Problem : Device turns ON when handled.  R4 (2.2MΩ) is too high impedant.  
* Proposed fix : 
    * Replace R4 by 100K || 100nF
    * Replace R9 by 100K (BoM simplification)
## TIR3
* Problem : Device can't be programmed.  On an ESP32-C3 IO2 and IO8 must be pulled high.
* Proposed fix :
    * add 10K pull up on IO2
    * add 10K pull up on IO8
## TIR4
* Problem : Excessive input current in sleep mode, caused by Q1 which had an internal diode from drain to gate. 
* Proposed fix : replacing the part by another SI3401A brings the current down to 200nA.

## Battery voltage measurement
This circuit consists of Q2, R6 and R5.

```python
# Measured data (input voltage on CN4 [V]: ADC-voltage on IO1 [mV])
voltage=[[5,1245],[4.9,1220],[4.8,1195],[4.7,1168],[4.6,1145],[4.51,1119], [4.41,1094],[4.31,1070], [4.21,1046], \
         [4.10,1019], [4.0,994], \
         [3.9,969],[3.8,945], [3.7,918], [3.61,893], [3.51,869], [3.40,845], [3.29,819], [3.20,791], [3.1,766], \
         [3.0,741], \
         [2.9,713], [2.8,687], [2.7,659]]
R6 = 10e3
R5 = 3.3e3
for measurement in voltage:
    vin_calc = measurement[1]/R5*(R5+R6)/1000
    error = (vin_calc - measurement[0])/vin_calc
    if abs(error) > 0.01:
        print("Measurement error is larger than 1% when Vin is {}V".format(measurement[0]))
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.title('Error of calculated voltage versus real measured voltage')
plt.xlabel('Measured voltage')
plt.ylabel('Error between calculated and measured voltage [%]')
x= [row[0] for row in voltage]
y= [measurement[1]/R5*(R5+R6)/1000 for measurement in voltage]
z = [(x1 - y1)/ x1 *100 for (x1, y1) in zip(x, y)]
plt.plot(x, z, color = "red", linewidth = 1.5, linestyle = "-.")
plt.show()
```

Remark that the ESP32-C3 is still working even though the input voltage is only 2.7V.  The Li-Ion battery should never be discharged this far.

## Antenna matching
### Procedure
1. NanoVNA stimulus setting : 768Mhz - 968MHz
2. NanoVNA calibration : soldering short, open, load on the BRD220522
3. Show the VSWR curve and check if the point of lowest VSWR corresponds to 868MHz.  If not, you might have to lengthen or shorten the antenna.  The actual impedance is very dependent on the location of your finger/hand, but that doesn't influence the frequency of lowest VSWR much, only the value of the lowest VSWR.  It's really important that the measurement is done with the PCB in the user's hand.  Covering the antenna by fingers has a huge influence on the impedance.
