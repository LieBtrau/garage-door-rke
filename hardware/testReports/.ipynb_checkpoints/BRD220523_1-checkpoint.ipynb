## Test incident reports

### TIR1
* Problem: Power supply does not start because the load is so high that it prevents the output voltage from rising.
* Action: Add an output switch which only connects the PSU to the load once the output voltage is high enough.

### TIR2
* Problem: Input filtering likely to be inadequate
* Action: Add two sections of 10µH/10µF

## TIR3
* Problem : Excessive ringing at low input voltages and high load.  Probably due to input capacitor grounds not directly connected to PCB ground, but through 2x via groups and ground plane.
* Action:
    * Add 10ohm in series with the boost capacitor
    * Add a snubber, calculated according to [Power Tips: Calculate an R-C snubber in seven steps](https://e2e.ti.com/blogs_/b/powerhouse/posts/calculate-an-r-c-snubber-in-seven-steps), derived from [Minimizing Ringing at the Switch Node of a Boost Converter](https://www.ti.com/lit/an/slva255/slva255.pdf) and [Snubber Circuits Suppress Voltage Transient Spikes in DC-DC Flyback Converter Supplies](https://www.maximintegrated.com/en/design/technical-documents/tutorials/8/848.html)
